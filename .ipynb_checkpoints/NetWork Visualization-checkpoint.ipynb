```python
def unstack(list1):
    unstack_list = [] 
    for x in list1: 
        if len(x) == 1:
            unstack_list.append(x) 
        if len(x) == 2:
            unstack_list.append(x[0]) 
            unstack_list.append(x[1]) 
    return unstack_list
```

```python
def unique(list1): 
    unique_list = [] 
    for x in list1: 
        if x not in unique_list: 
#             if len(x) == 1:
                unique_list.append(x) 
#             if len(x) == 2:
#                 unique_list.append(x[0]) 
#                 unique_list.append(x[1]) 
    return unique_list
```
